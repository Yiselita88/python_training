#series as specialized dictionary

```python
population_dict = {'California': 3833251, 'Texas': 26448193, "New York": 19651127, 'Florida': 19552860, 'Illinois': 12882135}
population_dict
population = pd.Series(population_dict)
population
```

```python
population['California':'Illinois']
```

```python
pd.Series([2, 4, 6])
```

```python
pd.Series(5, index=[100, 200, 300])
```

```python
pd.Series({2:'a', 1:'b', 3:'f', 6:'d', 4:'c', 5:'e'})
```

```python
pd.Series({2:'a', 1:'b', 3:'f', 6:'d', 4:'c', 5:'e'}).sort_index()
```

```python
pd.Series({2:'a', 1:'b', 3:'f', 6:'d', 4:'c', 5:'e'}).sort_values()
```

```python
pd.Series({2:'a', 1:'b', 3:'f', 6:'d', 4:'c', 5:'e'}).sort_index(ascending=False)
```

```python
pd.Series({2:'a', 1:'b', 3:'f', 6:'d', 4:'c', 5:'e'}).sort_values(ascending=False)
```
