# DataFrame as dictionary

```python
area = pd.Series({'California': 423967, 'Texas': 695662,
                  'New York': 141297, 'Florida': 170312,
                  'Illinois': 149995})
pop = pd.Series({'California': 38332521, 'Texas': 26448193,
                 'New York': 19651127, 'Florida': 19552860,
                 'Illinois':12882135})
data = pd.DataFrame({'area':area, 'pop':pop})
data
```

```python
data['area']
```

My data examples, family birthdays

```python
family_bday = pd.Series({'Mom': '10/28', 'Dad': '01/18', 'Dito': '10/11', 'husband': '01/05', 'MIL':'06/11', 'FIL': '09/13'})
family_age = pd.Series({'Mom': 59, 'Dad': 63, 'Dito': 31, 'husband': 36, 'MIL': 65, 'FIL': 68})
```

```python
new_data = pd.DataFrame({'bday':family_bday, 'age':family_age})
new_data
```

```python
new_data.loc['Mom']
```
